```python
# Section: Dictionaries (dict)
#
# Mutable        --> key-value pairs can be added, changed, or removed after creation.
# Ordered        --> since Python 3.7+, dicts maintain insertion order.
# Keys must be Hashable --> keys must be immutable (strings, ints, tuples with hashable elements).
# Values can be Anything --> values can be any Python object, including lists, dicts, functions.
# No Duplicate Keys --> assigning to an existing key OVERWRITES the value.
#
# Internally, Python implements dicts as hash tables.
# This gives O(1) average-case lookups, inserts, and deletes.
#
# IMPORTANT: Only the key must be hashable, not the value.
#   {[1,2]: 'a'}  --> TypeError: unhashable type: 'list'  (list as KEY)
#   {'a': [1,2]}  --> perfectly valid                     (list as VALUE)
#
# When to use dict:
#   - When you need fast lookup by a meaningful key (name, ID, code).
#   - When data is naturally in key → value pairs (JSON, configs, frequency maps).
#   - When you want to avoid repeated linear searches through a list.
```

```python
# a dict maps unique keys to values; keys must be hashable.

# Operations / concepts covered:
# 1.  Creation (multiple ways)
# 2.  Accessing values (key lookup & get())
# 3.  Adding & updating entries
# 4.  Deleting entries (del, pop(), popitem(), clear())
# 5.  keys(), values(), items()
# 6.  Membership testing (in, not in)
# 7.  update()
# 8.  setdefault()
# 9.  Dictionary comprehension
# 10. Nested dictionaries
# 11. Merging dicts (| operator, Python 3.9+)
# 12. Counting with dict (frequency map)
# 13. Sorting a dict by key / by value
# 14. defaultdict (collections)
# 15. Counter (collections)
# 16. OrderedDict (collections)
# 17. Unpacking with **
# 18. Shallow copy vs deep copy
# 19. Built-in functions: len
# 20. Iteration
```

# Creation

```python
d1 = {'name': 'Alice', 'age': 25}   # literal syntax
d1
```

```python
d2 = dict(name='Bob', age=30)       # using dict() with keyword args
d2
```

```python
d3 = dict([('x', 10), ('y', 20)])   # from list of (key, value) tuples
d3
```

```python
d4 = {([('x', 10), ('y', 20)])}
d4

# {} → Dictionary (if using key: value) or Set (if using values only).
```

```python
keys   = ['a', 'b', 'c']
values = [1, 2, 3]
d4 = dict(zip(keys, values))        # from two parallel lists
d4
```

```python
d5 = dict.fromkeys(['p', 'q', 'r'], 0)  # all keys share the same default value
d5
```

```python
d6 = {}          # empty dict
d7 = dict()      # also empty dict
print(d6, d7)
```

# Accessing Values

```python
student = {'name': 'Carol', 'grade': 'A', 'score': 95}

print(student['name'])               # 'Carol'  -- KeyError if key missing
print(student.get('grade'))          # 'A'
print(student.get('age', 'N/A'))     # 'N/A'   -- default if key missing, NO error
```

# Adding & Updating Entries

```python
config = {'host': 'localhost', 'port': 8080}

config['debug'] = True       # add new key
config['port']  = 9090       # update existing key
config
```

# Deleting Entries

```python
data = {'a': 1, 'b': 2, 'c': 3, 'd': 4}

del data['a']                  # remove key 'a'
print(data)                    # {'b': 2, 'c': 3, 'd': 4}

val = data.pop('b')            # removes 'b' and RETURNS its value
print(val)                     # 2
print(data)                    # {'c': 3, 'd': 4}

last = data.popitem()          # removes & returns LAST inserted (key, value)
print(last)                    # ('d', 4)

data.clear()                   # removes ALL items
print(data)                    # {}
```

# keys(), values(), items()

```python
person = {'name': 'Dave', 'age': 40, 'city': 'Delhi'}

print(list(person.keys()))    # ['name', 'age', 'city']
print(list(person.values()))  # ['Dave', 40, 'Delhi']
print(list(person.items()))   # [('name', 'Dave'), ('age', 40), ('city', 'Delhi')]
```

# Membership Testing

```python
info = {'brand': 'Toyota', 'model': 'Camry'}

print('brand' in info)         # True  -- checks KEYS only
print('brand' in info.keys()) 
print('Toyota' in info)        # False -- 'Toyota' is a VALUE, not a key
print('Toyota' in info.values())  # True
```

# update()

```python
base   = {'x': 1, 'y': 2}
extras = {'y': 99, 'z': 3}    # 'y' exists in base -- will be overwritten

base.update(extras)
base
```

# setdefault()

```python
cache = {'result': 42}

# if key EXISTS  --> returns existing value, does NOT change it
print(cache.setdefault('result', 0))   # 42

# if key MISSING --> inserts with default and returns it
print(cache.setdefault('status', 'ok'))  # 'ok'
print(cache)
```

# Dictionary Comprehension

```python
# {key_expr: value_expr for item in iterable if condition}

squares = {n: n**2 for n in range(1, 6)}
print(squares)    # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# swap keys and values
original = {'a': 1, 'b': 2, 'c': 3}
inverted = {v: k for k, v in original.items()}
print(inverted)   # {1: 'a', 2: 'b', 3: 'c'}

# filter: keep only high scorers
scores  = {'Alice': 85, 'Bob': 55, 'Carol': 92}
passing = {k: v for k, v in scores.items() if v >= 60}
print(passing)    # {'Alice': 85, 'Carol': 92}
```

# Nested Dictionaries

```python
company = {
    'employees': {
        'E001': {'name': 'Alice', 'dept': 'Engineering'},
        'E002': {'name': 'Bob',   'dept': 'Marketing'}
    },
    'founded': 2010
}

print(company['employees']['E001']['name'])   # 'Alice'
company['employees']['E001']['salary'] = 90000  # add nested key
company['employees']['E001']
```

# Merging Dicts (Python 3.9+ | operator)

```python
defaults = {'color': 'blue', 'size': 'M', 'weight': 1.5}
custom   = {'color': 'red', 'size': 'XL'}           # overrides defaults

merged = defaults | custom        # right side wins on conflict
print(merged)

# |= updates in place
defaults |= custom
print(defaults)
```

# Counting with Dict (Frequency Map)

```python
words = ['apple', 'banana', 'apple', 'cherry', 'banana', 'apple']

freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1

print(freq)   # {'apple': 3, 'banana': 2, 'cherry': 1}
```

# Sorting a Dict

```python
prices = {'banana': 1.2, 'apple': 0.8, 'cherry': 3.5, 'date': 2.1}

# sort by KEY
by_key = dict(sorted(prices.items()))
print(by_key)

# sort by VALUE (ascending)
by_val = dict(sorted(prices.items(), key=lambda item: item[1]))
print(by_val)

# sort by VALUE (descending)
by_val_desc = dict(sorted(prices.items(), key=lambda item: item[1], reverse=True))
print(by_val_desc)
```

# defaultdict (collections)

```python
from collections import defaultdict

# auto-initializes missing keys with the factory function
group = defaultdict(list)

students = [('Math', 'Alice'), ('Science', 'Bob'), ('Math', 'Carol')]
for subject, name in students:
    group[subject].append(name)   # no KeyError even on first access

dict(group)
```

# Counter (collections)

```python
from collections import Counter

text   = 'mississippi'
counts = Counter(text)
print(counts)                    # Counter({'s': 4, 'i': 4, 'p': 2, 'm': 1})
print(counts.most_common(2))     # [('s', 4), ('i', 4)]
```

# Unpacking with **

```python
def greet(name, age):
    print(f'Hello {name}, you are {age} years old.')

params = {'name': 'Eve', 'age': 28}
greet(**params)    # unpacks dict as keyword arguments

# combine dicts using **
a = {'x': 1}
b = {'y': 2}
merged = {**a, **b, 'z': 3}
print(merged)
```

# Shallow Copy vs Deep Copy

```python
import copy

orig  = {'a': [1, 2, 3], 'b': 42}
shallow = orig.copy()
deep    = copy.deepcopy(orig)

orig['a'].append(99)

print(shallow['a'])  # [1, 2, 3, 99]  -- inner list IS shared
print(deep['a'])     # [1, 2, 3]      -- fully independent
```

# Iteration

```python
record = {'id': 101, 'name': 'Frank', 'dept': 'HR'}

# iterate over keys (default)
for key in record:
    print(key, '->', record[key])

print()

# iterate over key-value pairs
for key, value in record.items():
    print(f'{key}: {value}')
```
