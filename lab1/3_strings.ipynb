```python
# Section: Strings (str)
#
# Immutable      --> once created, characters cannot be changed, added, or removed.
# Ordered        --> characters maintain their position; indexable by position.
# Indexable      --> individual characters can be accessed by index (positive and negative).
# Hashable       --> because strings are immutable, they can be used as dict keys or set elements.
# Allows Duplicates --> a string can contain repeated characters.
#
# Internally, Python stores strings as arrays of Unicode code points (UTF-8 / UCS-2 / UCS-4
# depending on the largest character in the string).
#
# Because strings are immutable, ALL string methods return a NEW string.
#   s = 'hello'
#   s.upper()   --> does NOT change s, returns a new string 'HELLO'
#
# IMPORTANT: Repeated concatenation in a loop is O(n^2).
#   Use ''.join(list_of_parts) for efficient string building.
#
# Strings vs bytes:
#   str  --> sequence of Unicode characters (text).
#   bytes --> sequence of raw bytes; for network, files, encoding.
```

### String interpolation (f-strings)

```python
name = "John"
age = 30
print(f"My name is {name} and I am {age} years old.")
```

### str.format()

```python
name = "John"
age = 50
print("My name is {} and I am {} years old.".format(name, age))
```

### % Operator

```python
name = "Johnathan"
age = 30
print("My name is %s and I am %d years old." % (name, age))
```

# Creation

```python
s1 = 'hello'                # single quotes
s2 = "world"                # double quotes (identical to single)
s3 = str(42)                # from int  --> '42'
s4 = str(3.14)              # from float --> '3.14'
s5 = str(['a', 'b'])        # from list  --> "['a', 'b']"
print(s1, s2, s3, s4, s5)
```

```python
# Triple quotes for multi-line strings
poem = """Roses are red,
Violets are blue,
Python is awesome,
And so are you."""
print(poem)
```

```python
# Raw string -- backslashes are treated literally (useful for file paths, regex)
path  = r'C:\Users\name\Documents'   # r prefix
print(path)
```

# Indexing

```python
s = 'Python'
#    0 1 2 3 4 5   (positive indices)
#   -6-5-4-3-2-1   (negative indices)

print(s[0])    # 'P'  -- first character
print(s[3])    # 'h'
print(s[-1])   # 'n'  -- last character
print(s[-3])   # 'h'
```

# Slicing

```python
s = 'Hello, World!'

print(s[0:5])     # 'Hello'   -- stop is exclusive
print(s[7:])      # 'World!'  -- to end
print(s[:5])      # 'Hello'   -- from start
print(s[::2])     # every 2nd character
print(s[::-1])    # reverse the string
```

# Immutability

```python
s = 'hello'
# s[0] = 'H'   --> TypeError: 'str' object does not support item assignment

# To 'modify', create a NEW string
s = 'H' + s[1:]    # 'Hello'
print(s)

# All methods also return new strings
original = 'hello'
upper    = original.upper()    # new string
print(original)                # 'hello'  -- unchanged
print(upper)                   # 'HELLO'
```

# len()

```python
print(len('hello'))        # 5
print(len(''))             # 0  -- empty string
print(len('hello world'))  # 11 -- spaces count
```

# Concatenation & Repetition

```python
first = 'Hello'
last  = 'World'

full  = first + ', ' + last + '!'   # concatenation -- creates new string
print(full)                          # 'Hello, World!'

line  = '-' * 20                     # repetition
print(line)

# Efficient multi-part concatenation
parts  = ['apple', 'banana', 'cherry']
result = ', '.join(parts)            # preferred over += in loops
print(result)
```

# Membership Testing

```python
s = 'The quick brown fox'

print('quick' in s)          # True  -- substring check
print('slow' not in s)       # True
print('Q' in s)              # False -- case-sensitive
```

# Case Methods

```python
s = 'hElLo WoRlD'

print(s.upper())       # 'HELLO WORLD'
print(s.lower())       # 'hello world'
print(s.title())       # 'Hello World'   -- each word capitalised
print(s.swapcase())    # 'HeLlO wOrLd'   -- flip case of each char
print(s.capitalize())  # 'Hello world'   -- only first char upper, rest lower
```

# Strip Methods

```python
s = '   hello world   '

print(repr(s.strip()))    # 'hello world'  -- both ends
print(repr(s.lstrip()))   # 'hello world   '  -- left only
print(repr(s.rstrip()))   # '   hello world'  -- right only

# strip specific characters
t = '###warning###'
print(t.strip('#'))       # 'warning'
```

# Find & Index

```python
s = 'banana'

print(s.find('an'))      # 1  -- index of FIRST occurrence (-1 if not found)
print(s.rfind('an'))     # 3  -- index of LAST occurrence
print(s.find('xyz'))     # -1 -- not found, NO error

# index() is like find() but raises ValueError if not found
print(s.index('an'))     # 1
# s.index('xyz')         --> ValueError

print(s.count('a'))      # 3  -- how many times 'a' appears
```

# Replace

```python
s = 'I love cats. Cats are great. Cats!'

print(s.replace('Cats', 'Dogs'))           # replace ALL occurrences
print(s.replace('Cats', 'Dogs', 1))        # replace only FIRST occurrence
print(s)                                   # original unchanged
```

# Split & Join

```python
sentence = 'one two three four five'

words = sentence.split()         # split on whitespace by default
print(words)                     # ['one', 'two', 'three', 'four', 'five']

csv_row = 'a,b,c,d'
print(csv_row.split(','))        # ['a', 'b', 'c', 'd']
print(csv_row.split(',', 2))     # ['a', 'b', 'c,d']  -- max 2 splits

# splitlines() splits on newline characters
multi = 'line1\nline2\nline3'
print(multi.splitlines())        # ['line1', 'line2', 'line3']
```

```python
parts = ['2024', '06', '15']
print('-'.join(parts))           # '2024-06-15'
print(' | '.join(parts))         # '2024 | 06 | 15'
print(''.join(parts))            # '20240615'
```

# startswith() & endswith()

```python
url  = 'https://www.example.com'
file = 'report_2024.pdf'

print(url.startswith('https'))         # True
print(file.endswith('.pdf'))           # True

# can pass a tuple of options
print(file.endswith(('.pdf', '.docx', '.xlsx')))  # True
```

# Check Methods

```python
print('Hello'.isalpha())    # True   -- only letters
print('123'.isdigit())      # True   -- only digits
print('Hello3'.isalnum())   # True   -- letters or digits
print('   '.isspace())      # True   -- only whitespace
print('HELLO'.isupper())    # True
print('hello'.islower())    # True
print('Hello World'.istitle())  # True
```

# String Formatting

```python
name  = 'Midori'
score = 98.765

# f-string (preferred, Python 3.6+)
print(f'Name: {name}, Score: {score:.2f}')   # 2 decimal places

# str.format()
print('Name: {}, Score: {:.2f}'.format(name, score))

# Named placeholders
print('{n} scored {s:.1f} points'.format(n=name, s=score))

# % operator (older style)
print('Name: %s, Score: %.2f' % (name, score))
```

```python
# f-string: numbers, alignment, padding
pi = 3.14159265
print(f'{pi:.4f}')        # 3.1416   -- 4 decimal places
print(f'{1000000:,}')     # 1,000,000 -- thousands separator
print(f'{0.85:.1%}')      # 85.0%     -- percentage
print(f'{42:08d}')        # 00000042  -- zero-padded
```

# Alignment: center(), ljust(), rjust(), zfill()

```python
s = 'Python'

print(s.center(20, '-'))   # '-------Python-------'
print(s.ljust(20, '.'))    # 'Python..............'
print(s.rjust(20, '.'))    # '..............Python'
print('42'.zfill(8))       # '00000042'  -- zero-fill to width
```

# Escape Sequences & Raw Strings

```python
# Common escape sequences
print('Hello\nWorld')     # newline
print('tab\there')        # tab
print('quote\'s')         # escaped single quote
print('back\\slash')      # literal backslash

# Raw string: no escape processing
import re
pattern = r'\d+\.\d+'    # match a decimal number  (raw is cleaner here)
print(re.findall(pattern, 'Price: 3.99 and 12.50'))
```

# String Unpacking & Iteration

```python
# Unpack characters
a, b, c = 'XYZ'
print(a, b, c)    # X Y Z

# Iterate character by character
for ch in 'hello':
    print(ch, end=' ')   # h e l l o
print()

# With enumerate
for idx, ch in enumerate('abc'):
    print(idx, ch)
```

# encode() & decode()

```python
s = 'hello'

# str --> bytes
b = s.encode('utf-8')       # b'hello'
print(b)
print(type(b))              # <class 'bytes'>

# bytes --> str
original = b.decode('utf-8')  # 'hello'
print(original)
print(type(original))         # <class 'str'>

# Unicode example
emoji = '\u2764 Python'
print(emoji)                  # ❤ Python
print(emoji.encode('utf-8'))  # bytes representation
```
