```python
# Section 5: Sets (set)
# 
# Mutable
# Not Hashable (because set is mutable)
# not Indexable
# not ordered

# python internally uses a hash table to store the elements into memory location.
# hash(element) --> location in memory
# like lists python does not scan elements one by one .. it directly jumps to memory location in O(1) TC

# elements inside a set must be hashable because sets are implemented as hash tables. This means that you cannot have mutable types 
# (like lists or dictionaries) as elements of a set, but you can have immutable types (like integers, strings, and tuples) as elements of a set.

# hash("Souvik") ---> 123458887 (memory location)
# hash((1,2,3)) ---> 1232940 (memory location)
# hash(10) ---> 124747890 (memory location)

# these are immutable types and they can be hashed, so they can be added to a set.


# ------------------------------------------------------------------------------------------------

# x = [1,2,3]
# set = {x} # TypeError: unhashable type: 'list'
# because we can do x.append(4) and change the value of x, so it is mutable and unhashable, so the memory location also changes

# Not all immutable objects are hashable.
# The set itself is not hashable because it is mutable, but frozenset is hashable because it is immutable.
```

```python
a = {1, 2}
b = {1, 2, 3}
c = {4, 5}
```

## Set Operations

```python
print(a | b )  # union
print(a & b )  # intersection
print(a - b)   # difference
print(a ^ b)   # symmetric difference
```

```python
# The above operators are more concise and easier to read, but the following methods can also be used to achieve the same results:

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
print(a.symmetric_difference(b))
```

## Set Relations

```python
# every set is a subset of itself and every set is a superset of itself.
```

```python
print(a <= b)   # subset            a.issubset(b) 
print(a < b)    # proper subset     a.issubset(b) and a != b
print(a >= b)   # superset          a.issuperset(b)
print(a > b)    # proper superset   a.issuperset(b) and a != b
```

```python
a.isdisjoint(c) # returns True if a and c have no elements in common, otherwise it returns False.
# These methods return booleans and are optimized because sets use hash tables internally.

# this is faster than 
len(a&c) == 0
# bacause it does not need to compute the intersection of a and c, it just needs to check if there is any common element between a and c.
```

## Copy

```python
# creats a shallow copy of the set
d = a.copy()
d
```

```python
d == a
```

```python
d is a # false because d is a new set object in memory, but it has the same elements as a, so d == a is true.
```

```python
# without copy
b = a
a.add(165)
b

# here the problem is that we have created a new variable b and assigned it the value of a, but since sets are mutable, both a and b point to the same set object in memory. 
# So when we add 165 to a, it also gets added to b because they are the same object in memory.

# if we create b using copy method as such b = a.copy() then both are different objects in memory and changing one will not affect the other.
```

```python
a = {1, 2}
b = a.copy()

b.add(3)

print(a) 
print(b) 
```

## FrozenSet

```python
s = set([1,2,3])
s
```

```python
fs = frozenset((1, 2, 3))
fs

# frozenset is
# 1. immutable
# 2. hashable
# frozenset is a built-in function that takes an iterable as an argument and returns an immutable set object.

# can be used as a key to dictionary because it is hashable
# it can be also used as an element to anothr set because it is hashable, 
# but a set cannot be an element to another set because it is not hashable.
```

```python
fs.add(4) # AttributeError: 'frozenset' object has no attribute 'add'
```

```python
ds = fs.union({4, 5}) # returns a new frozenset that is the union of fs and {4, 5}
ds
```

```python
fs
```

```python
hash(fs) # returns the hash value of the frozenset object, which is an integer that represents the memory location of the object.
```

```python
hash(ds)
```
