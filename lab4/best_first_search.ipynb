```python
import random

def generate_maze(rows, cols, wall_prob=0.3):
    maze = []
    for r in range(rows):
        row = []
        for c in range(cols):
            if random.random() < wall_prob:
                row.append(0)
            else:
                row.append(1)
        maze.append(row)

# This guarantees the algorithm can at least attempt a path search.
    maze[0][0] = 1
    maze[rows-1][cols-1] = 1
    return maze
```

```python
def print_maze(maze,path=None):
    path = set(path) if path else set()

    for r in range(len(maze)):
        for c in range(len(maze[0])):
            if (r,c) in path:
                print(" P ", end=" ")
            elif maze[r][c] == 0:
                print(" | ", end=" ")
            else:
                print(" . ", end=" ")
        print()
```

```python
rows = 12
cols = 12
maze = generate_maze(rows, cols, wall_prob=0.20)
directions = [(-1,0),(0,1),(1,0),(0,-1)]

start = (0,0)
end = (rows-1,cols-1)
```

```python
# BFS + Priority Queue + Heuristic
# Queue → explore level by level
# Priority Queue → explore most promising node first

# h(n) = Manhattan distance to treasure
# (r1,c1) and (r2,c2)
# |r1 - r2| + |c1 - c2|

# Start: (1,2)
# Treasure: (4,5)
# |1-4| + |2-5|
# = 3 + 3
# = 6
```

```python
import heapq

def best_first_search(maze,start,end):

    rows = len(maze)
    cols = len(maze[0])
    directions = [(-1,0),(1,0),(0,-1),(0,1)]

    pq = []
    heapq.heappush(pq,(0,start))

    visited = {start}
    parent = {}

    while pq:
        h,(r,c) = heapq.heappop(pq)

        if (r,c) == end:
            path = []
            node = end

            while node != start:
                path.append(node)
                node = parent[node]
            path.append(start)
            path.reverse()
            return path


        for dr,dc in directions:
            nr = r + dr
            nc = c + dc

            if 0 <= nr < rows and 0 <= nc < cols:
                if maze[nr][nc] == 1 and (nr,nc) not in visited:
                    visited.add((nr,nc))
                    parent[(nr,nc)] = (r,c)
                    h = abs(nr-end[0]) + abs(nc-end[1])   # Manhattan distance
                    heapq.heappush(pq,(h,(nr,nc)))
    return None
```

```python
path_best = best_first_search(maze,start,end)

if path_best:
    print("\nBest First Search path length:",len(path_best)-1)
    print_maze(maze,path_best)
```
