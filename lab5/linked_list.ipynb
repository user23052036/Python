```python
class Node:
    def __init__(self, val=None, next=None):
        self.val=val
        self.next=next
```

```python
class LinkedList:
    def __init__(self):
        self.head=None
    
    def insert_at_beginning(self,val):  # O(1)
        new_head = Node(val,self.head)
        self.head = new_head
    
    def insert_at_end(self,val):  # O(n)
        if self.head is None:
            self.head = Node(val)
            return
        
        tail = Node(val)
        temp = self.head

        while temp.next != None:
            temp = temp.next
        temp.next = tail
    
    def size(self):
        if self.head is None:
            return 0
        cnt=0
        temp = self.head
        while temp:
            cnt += 1
            temp = temp.next
        return cnt
    
    def insert_at_position(self,val,pos): 
        """assuming 1-indexed like array"""
        if self.head is None and pos != 1:
            return
        if pos == 1:
            self.insert_at_beginning(val)
        
        if pos == self.size()+1:
            self.insert_at_end(val)
        else:
            temp = self.head
            for _ in range(1,pos-1):
                temp = temp.next
                pos += 1
            new_node = Node(val)
            next_node = temp.next

            temp.next = new_node
            new_node.next = next_node

    def list_to_linkedlist(self,data_list):
        for item in data_list:
            self.insert_at_end(item)

    def print(self):
        if self.head is None:
            print("we have no LinkedList")
            return
        
        temp = self.head
        while temp!=None:
            print(temp.val,'--> ',end='')
            temp = temp.next
```

```python
lst = LinkedList()
lst.insert_at_beginning(12)
lst.insert_at_beginning(32)
lst.insert_at_beginning(3)
lst.insert_at_end(89)
lst.insert_at_end(100)
lst.print()
```

```python
fruits = ['mango','banana','cucumber','carrot','tomato','capsicum']
lst1 = LinkedList()
```

```python
lst1.list_to_linkedlist(fruits)
lst1.print()
```

```python
lst2 = LinkedList()
lst2.insert_at_beginning(3)
lst2.insert_at_beginning(2)
lst2.insert_at_beginning(1)
lst2.print()
```

```python
lst2.insert_at_position(40,3)
lst2.print()
```
