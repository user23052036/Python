```python
def change(L):
    print(id(L))
    L.append(5)
    print(id(L))

L1 = [1,2,3,4]
print(id(L1))
print(L1)

change(L1[:])   # important
print(L1)

# L1[:] creates a new copy of the list.
# L1 → original list
# L1[:] → new list (different memory address)

"""
L1  ──► [1,2,3,4]
L1[:] ──► [1,2,3,4]   (new box, same contents)
"""
```
