Python technically uses **"pass by object reference,"** but you can achieve the effect of **pass by value** by passing a **copy** of the object. This ensures the original object remains unchanged even if the function modifies the input.

---

### 1. For Simple Lists or Dictionaries

Use the `.copy()` method or slicing `[:]`.

```python
# Passing a copy of a list
print_id(cust.copy()) 

# Or using slicing
print_id(cust[:])

```

### 2. For Deep/Nested Objects

If your object contains other objects (like a list inside a list), a simple copy isn't enough. You need `deepcopy`.

```python
import copy

# Creates a brand new object with its own memory address
print_id(copy.deepcopy(cust))

```

### 3. For Custom Objects (Classes)

If `cust` is an instance of a class you wrote, you can use the `copy` module as shown above, or manually pass the attributes.

```python
# Passing attributes to a new instance (Manual pass by value)
print_id(Customer(cust.name, cust.age))

```

---

### Why `id()` changes when you do this

Internally, when you "pass by value" in Python, you are actually creating a **new object in memory** with the same data.

| Method | Effect | Internal Behavior |
| --- | --- | --- |
| **Direct Pass** | Pass by Reference | `customer` and `cust` point to the same memory address. |
| **`.copy()`** | Pass by Value (Shallow) | Creates a new outer object; `id()` will be different. |
| **`deepcopy()`** | Pass by Value (Deep) | Creates a new outer object and copies all nested objects inside it. |


The reason you are seeing these errors is that `.copy()` and `[:]` are built-in methods for collections like **lists** and **dictionaries**, but they do not exist for **custom classes** unless you define them.

### 2. Why your attempts failed

* **`cust.copy()`**: Your `CUSTOMER` class does not have a method named `copy`. Python does not automatically add this to user-defined classes.
* **`cust[:]`**: Slicing is for sequences (lists, strings). Since `CUSTOMER` doesn't define how to handle indexes, it is "not subscriptable."

---

### Internal Comparison

| Approach | What happens internally | Resulting `id()` |
| --- | --- | --- |
| `print_id(cust)` | Points to the **original** memory address. | **Same** |
| `print_id(copy.copy(cust))` | Creates a **new object** and copies attributes. | **Different** |
| `print_id(CUSTOMER(cust.name...))` | Runs `__init__` again to build a **new object**. | **Different** |

````python
def change(L):
    print(id(L))
    L.append(5)
    print(id(L))

L1 = [1,2,3,4]
print(id(L1))
print(L1)

change(L1[:])   # important
print(L1)

# L1[:] creates a new copy of the list.
# L1 → original list
# L1[:] → new list (different memory address)

"""
L1  ──► [1,2,3,4]
L1[:] ──► [1,2,3,4]   (new box, same contents)
"""
````
