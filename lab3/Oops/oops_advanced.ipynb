## Relationship in Python: Aggregation vs. Inheritance

### 1. Inheritance ("Is-A" Relationship)

inheritance represents a specialized version of a general category.

* **Concept:** A child class takes on all the attributes and methods of a parent class.
* **Examples from your notes:** * A **Smartphone** is-a **Product**.
* A **Car** is-a **Vehicle**.


* **Why use it?** To avoid code duplication. If every `Vehicle` has a `fuel_type`, you define it once in the parent rather than in `Car`, `Truck`, and `Bike` separately.

### 2. Aggregation ("Has-A" Relationship)

illustrates this using the `Customer` and `Address` classes.

* **Concept:** One class contains an instance of another class as one of its attributes. The classes are independent; an `Address` can exist without a `Customer`.
* **Example from your code:**
* A **Customer** has-an **Address**.


* **Implementation:** In `customer.py`, you pass the `address` object into the `CUSTOMER` constructor:
`self.address = address`
* **Why use it?** It creates a "modular" design. You can change how an `Address` works without breaking the `CUSTOMER` class.

## Addressing Your Errors

The `TypeError` you are seeing is a very common mistake when working with "Magic Methods" like `__repr__` or `__str__`.

### The Problem

In both your `CUSTOMER` and `Address` classes, you used **`print()`** inside the `__repr__` method:

```python
def __repr__(self):
    print(f"name: {self.name}...") 

```

**Why it fails:** Python's internal machinery expects `__repr__` to **`return`** a string so that it can display it in the console.

1. The `print()` function performs its action but returns `None`.
2. When you type `customer1` in your notebook, Python calls `repr(customer1)`.
3. Because your method returns `None` (via `print`), Python throws: `TypeError: __str__ returned non-string (type NoneType)`.

### The Fix

Change **`print`** to **`return`** in both classes.

#### Corrected `Address` class:

```python
def __repr__(self):
    return f"city: {self.city} pin: {self.pincode} state: {self.state}"

```

#### Corrected `CUSTOMER` class:

```python
def __repr__(self):
    # Note: Calling self.address here works because Address now returns a string!
    return f"name: {self.name}\n gender: {self.gender}\n address: {self.address}"

```

### Summary of Changes

| Feature | Your Current Code | Corrected Code |
| --- | --- | --- |
| **Action** | `print(f"...")` | `return f"..."` |
| **Return Type** | `NoneType` | `String` |
| **Result** | Crashes the REPL/Notebook | Displays object details correctly |

## Super keyword

````python
class Phone:
    def __init__(self,brand,battery,camera):
        self.__brand = brand 
        self.battery = battery
        self.camera = camera
        print('Inside phone class constructor')
    
    def buy(self):
        print('buying a phone')

class SmartPhone(Phone):

    def __init__(self,brand,battery,camera,ram,os):
        super().__init__(brand,battery,camera) # parent constructor got invoked
        self.ram = ram
        self.os = os
        print('Inside smartphone class constructor')

    def buy(self):
        super().buy()  # parent ka buy method got invoked
        print('buying a smartphone')
````

````python
s = SmartPhone('oppo','5000',50,1,2)
s.buy()
````

````python
s.super().buy() # error because super keyword class k ahar kam nhi karta
# only parent method and constructors can be accessed not even data_members
````

````python
# 4 types of inheritence:

# 1. single level inheritence
# 2. multi level inheritence  parent---> child----> grandchild
# 3. hirarchical level inheritence      parent ---->child1
                                     #     |
                                     #     |
                                     #     V
                                     #   child2
# 4. multiple ingeritence not possible in java but possible in python 
#    child inherites properties of more than 1 parents
# 5. hybrid inheritence (combinations of above inheritence)
````

````python
# multi level inheritence (chain inheritence)

class A:
    def m1(self):
        return 20
    
class B(A):
    def m1(self):
        return 30
    def m2(self):
        return 40
    
class C(B):
    def m2(self):
        return 20
````

````python
obj1 = A()
obj2 = B()
obj3 = C()

print(obj1.m1() + obj3.m1() + obj3.m2())
````
