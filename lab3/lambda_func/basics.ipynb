```python
# lambda function is used with heigher order function
# heigher order function is defined as function whose input is another function

numbers = [12,4,55,2,13,90,100,19,3]

x = lambda x: x%2==0
y = lambda x: x%2!=0
z = lambda x: x%3==0

#HOF(heigher order function)
def return_sum(func,L):
    result=0

    for i in L:
        if func(i):
            result += i
    return result
```

```python
print(return_sum(x,numbers)) # === print(return_sum(lambda x: x%2==0,numbers))
print(return_sum(y,numbers))
print(return_sum(z,numbers))
```
