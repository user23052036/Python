### Reduce

```python
import functools
```

```python
L = [1,2,3,4,5,6,7,8]
```

```python
functools.reduce(lambda x,y: x+y, L)
```

```python
functools.reduce(lambda x,y: x if x>y else y, L)
```

The primary difference lies in **how** the function is applied to your data. While a simple `lambda` is just a tool (a small function), `map`, `filter`, and `reduce` are the "engines" that drive that tool across a collection.

---

### 1. Map: The Transformer

`map()` applies the lambda function to **every** item in the list and returns a new collection of the same length.

* **Internal Logic:** It creates a loop behind the scenes, takes one element at a time, passes it to the lambda, and stores the result.
* **Result Size:** Always the same length as the input.
* **Analogy:** You have a list of people and you give **everyone** a hat.

### 2. Filter: The Gatekeeper

`filter()` uses the lambda function to test each item. The lambda must return `True` or `False`.

* **Internal Logic:** It loops through the list and calls the lambda. If the lambda returns `True`, the item stays; if `False`, it’s discarded.
* **Result Size:** Usually smaller than the input (or empty).
* **Analogy:** You have a list of people and you only let those **over 18** into a club.

### 3. Reduce: The Aggregator

`reduce()` (from `functools`) "shrinks" the entire list into a **single value**.

* **Internal Logic:** It takes the first two elements, applies the lambda, takes that result and the third element, applies the lambda again, and repeats until only one value remains.
* **Result Size:** Always exactly one value.
* **Analogy:** You have a list of numbers and you **add them all up** into one total sum.

---

### Comparison Table

| Function | Purpose | Lambda Requirement | Output |
| --- | --- | --- | --- |
| **Map** | Transformation | Perform an action on `x` | List of new values |
| **Filter** | Selection | Return a boolean (True/False) | Subset of original list |
| **Reduce** | Accumulation | Combine `x` and `y` | A single value |
