````python
"""
encapsulation

object.pin and object.balance is accessible and can be modified
to prevent this we had access modifiers in java like public,private,protected
in python we have __ (double underscore attached infront of pin and balance === private)
works for both data and methods

__pin ----> is internally created as _ATM__pin
__balance ---> _ATM__balance

unfortunaely we can do sbi._ATM__balance = 'dfw74vdv`d'
nothing is PRIVATE IN PYTHON !

Encapsulation in Python is the practice of **binding data and the methods that operate on it inside a class, 
restricting direct access to internal attributes, and providing controlled access through getter and setter 
methods to view or modify constructor data using custom logic.**

step-1: identify the private attribute (__)
step-2: define geeter and setter methods using custom logic

"""
````

````python
# Class Diagram

"""
+-----------------------------+
|             ATM             |
+-----------------------------+
| - __pin : str               |
| - __balance : float         |
+-----------------------------+
| + __init__()                |
| + menu()                    |
| + set_pin(new_pin: str)     |
| + get_pin() : str           |
| + get_balance() : float     |
| + deposit()                 |
| + withdraw()                |
| + check_balance()           |
+-----------------------------+


In Python:

__something → name mangled (private-like)
_something → protected (convention)
__something__ → special system method (public, not private)
"""
````


#### The Real Difference

Without control:

```python
atm.__balance = 99999999   # (if it were public)
```

This overwrites state blindly.

With `_update_balance`:

* Only deposit/withdraw call it.
* Business logic checks happen first.
* No arbitrary overwrite.
* Balance changes only via delta (+ / -).

So control = logic layer between input and state mutation.

#### Hard Truth

If someone wants to break your class:

```python
atm._ATM__balance = 100000000
```

They can.

Encapsulation in Python protects against **accidental misuse**, not malicious actors.
