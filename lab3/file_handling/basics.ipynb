### Some Theory

##### Types of data used for I/O:
- Text - '12345' as a sequence of unicode chars
- Binary - 12345 as a sequence of bytes of its binary equivalent

##### Hence there are 2 file types to deal with
- Text files - All program files are text files
- Binary Files - Images,music,video,exe files

### How File I/O is done in most programming languages

- Open a file
- Read/Write data
- Close the file

write() returns characters written because Python follows the POSIX write semantics, where write operations report the number of bytes written so programs can detect partial writes.

```python
# case 1 - if the file is not present
f = open('sample.txt','w')
n1 = f.write('Hello world')
n2 = f.write('\n how the hell are u >?')
f.close()

# f.close()
# since file is closed hence this will not work
# f.write('hello')

# close() does two things:
# 1. Flushes the buffer (forces pending writes to disk)
# Releases the file descriptor
```

```python
print(n1+n2)
```

```python
# appending in existing file
f = open('sample.txt','a')
f.write('\nhola milamo sonu! mucho gusto !!!')
f.close()
```

```python
# write multiple lines
L = ['\nkito','\nmiuy','\nkung-fu','\nquando-tro','\nzoraa']

f = open('sample.txt','w')
f.writelines(L)
f.close()
```

```python
# reading upto n chars
f = open('sample.txt','r')
s = f.read(10)
print(s)
f.close()
```

```python
# readline() --> reading line by line
# by default print and readline() both adds nextine charecter so we stop one
f = open('sample.txt','r')
print(f.readline(),end='')
print(f.readline(),end='')
f.close()
```

```python
# custom loop to print content using readline
f = open('sample.txt','r')

while True:
    line = f.readline()
    if line == '':
        break
    else:
        print(line,end='')
f.close()
```

### Context manager(WITH)
- if we forget to close the opened file the garbage collector will eventually close
- with helps us not to write f.close() as it closes by its own when the usages is over

```python
with open('sample.txt','a') as f:
    f.write('\ndoes this work!!!')
```

```python
# print next 10 charecters too
with open('sample.txt','r') as f:
    f.read(10)
    print(f.read(10))
```

```python
# loading big_file in chunks
big_L = ['niga ho me dekho meri jo he bas gaya!! ... wo he milta tumse hubhueee..\n' for i in range(500)]
with open('bigfile.txt','w') as f:
    f.writelines(big_L)
```

```python
with open('bigfile.txt','r') as f:
    chunk_size = 50
    while True:
        chunk = f.read(chunk_size)
        if not chunk:
            break
        print(chunk, end='\n\n')
```
